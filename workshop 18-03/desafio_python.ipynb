```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# --- 1. NUMPY ---
# 7 temperaturas aleatórias entre 20°C e 35°C
temperaturas = np.random.uniform(20, 35, 7)
dias_semana = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sáb', 'Dom']
print("Dias da semana: ", dias_semana)
print("Temperaturas: ", [f"{temp:.2f}" for temp in temperaturas])
```

```python
# --- 2. PANDAS ---
df = pd.DataFrame({
    'Dia': dias_semana,
    'Temperatura': temperaturas
})

# Calculos
media_temp = df['Temperatura'].mean()
temp_max = df['Temperatura'].max()
temp_min = df['Temperatura'].min()

print(f"Média da semana: {media_temp:.2f}°C")
print(f"Máxima: {temp_max:.2f}°C | Mínima: {temp_min:.2f}°C")
```

```python
# --- 3. MATPLOTLIB ---
# Uma figura com 3 subplots para mostrar variações diferentes
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Gráfico 1 - Linhas
axes[0].plot(df['Dia'], df['Temperatura'], marker='o', color='orange', linestyle='-')
axes[0].set_title('Variação Diária (Linha)')
axes[0].set_ylabel('Temperatura (°C)')

# Gráfico 2 - Barras
axes[1].bar(df['Dia'], df['Temperatura'], color='blue')
axes[1].set_title('Comparação por Dia (Barras)')

# Gráfico 3 - Área
axes[2].fill_between(df['Dia'], df['Temperatura'], color='lightgreen', alpha=0.4)
axes[2].plot(df['Dia'], df['Temperatura'], color='green', marker='s')
axes[2].set_title('Intensidade Térmica (Área)')

plt.tight_layout()
plt.show()
```
